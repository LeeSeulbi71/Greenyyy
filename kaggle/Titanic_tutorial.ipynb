**Pandas 기능**

1. 간단 산술 통계

  - info(): 데이터들의 정보 열람
  - describe(): 데이터의 산술 통계 정보 (산술 가능 = 숫자로 표현 가능)
  - isnull(): 데이터 누락?
  - sum(): 데이터 합산
  - value_ counts: 데이터 종류별 갯수 확인

2. 고급 산술 통계
  - groupby ( = pivot_table): 특정 column 데이터 

3. 조건 필터링
  - loc: 조건문에 대한 필터를 걸어 데이터 확인
  - &와 | 연산자: & and , | or

4. 데이터 전처리
  - replace(): 데이터를 다른 값으로 바꿀 때
  - fillna(): 빈 값 채우기 (Na = Not a Number)
  - map(): 데이터 매핑
